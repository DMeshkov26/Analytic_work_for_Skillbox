```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_courses = pd.read_csv('courses.csv')
df_course_contents = pd.read_csv('course_contents.csv')
df_progresses = pd.read_csv('progresses.csv')
df_progress_phases = pd.read_csv('progress_phases.csv')
df_students = pd.read_csv('students.csv', parse_dates=['birthday'], dayfirst=False)
df_courses = df_courses.rename(columns={'id': 'course_id'})
df_progresses = df_progresses.rename(columns={'id': 'progress_id'})
df2 = pd.merge(df_courses, df_course_contents)
df1 = pd.merge(df_progresses,df_progress_phases)
df = pd.merge(df2,df1)
df['start_date'] = df['start_date'].astype('datetime64')
df['finish_date'] = df['finish_date'].astype('datetime64')
```

```python
# Задание 1

df_f = df[df['is_homework']==True].groupby(['title', 'student_id'])['start_date'].min().reset_index()
dff = pd.merge(df_f, df)
dff['start_date'] = df_f['start_date'].astype('datetime64[M]').dt.to_period('M')
data = dff.groupby(['title', 'start_date'])['student_id'].count().reset_index().sort_values(['title', 'start_date'])
data
```

```python
# Задание 2

fig, axes = plt.subplots(figsize = (15,25), ncols = 3, nrows = 5, sharey = True)
titles = data['title'].unique()
fig.suptitle("Прирост учеников на разных курсах по месяцам", y = 1)
for index, ax in enumerate(axes.flat):
    plot_df = data[data['title']==titles[index]].pivot_table('student_id', index='start_date')
    plot_df.plot(ax=ax, title=titles[index], legend=False)
    ax.set_xlabel('Дата начала прохождения курса')
    ax.set_ylabel('Количество студентов')

fig.tight_layout()
plt.show()
```

1. **(p)** Посчитать конверсию перехода студентов из одного модуля в другой на каждом курсе. <details>

3. **(m)** Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. 15 графиков. Графики должны иметь заголовки. Ticks должны содержать номер и название модуля. Цвет столбцов графиков должен содержать информацию о сфере, к которой относится курс (нужна легенда). Прозрачность должна стоять на отметке 0.1. На графики должна быть нанесена линия медианы конверсии для каждого курса. У медианы должен быть свой цвет. Рамок у графиков быть не должно ⭐
4. На основании рассчитанных значений сделайте аналитический вывод (должен быть полным и покрывать все полученные выше метрики).

```python
# Задание 1

conversion_df = df.loc[df['is_homework']==True, 
    ['title', 'module_number', 'field','module_title', 'student_id', 'status','start_date', 'finish_date']]

conversion_df = conversion_df.assign(
    start = conversion_df['start_date'].notnull(), finish=conversion_df['finish_date'].notnull()).groupby(
    ['title', 'field','module_number', 'module_title']).agg({'start' : np.count_nonzero, 'finish' : np.count_nonzero}).reset_index()

conversion_df['conversion'] = '0'

titles = conversion_df['title'].unique() 

done_df = pd.DataFrame(
    {'title':[], 'field': [], 'module_number': [], 'module_title': [], 'start': [], 'finish': [], 'conversion': []})

for title in titles:
    convers_title = conversion_df[conversion_df['title']== title].reset_index()
 
    for i in range(convers_title.index.max() +1):
        if i == 0:
            convers_title.loc[i, 'conversion'] = 1
        else:
            convers_title.loc[i, 'conversion'] = convers_title.loc[i, 'start'] / convers_title.loc[i-1, 'finish'] if convers_title.loc[i-1, 'finish'] > 0 else 0 
            
        
    done_df = pd.concat([done_df, convers_title], ignore_index=True)

done_df['conversion'] = done_df['conversion'].apply(pd.to_numeric)  

done_df.head(20)
```

```python
# Задание 2

fig, ax = plt.subplots(figsize=(24, 12))

done_df.pivot_table(values='conversion', index=done_df.index, columns='title').plot.bar(
    ax=ax, title='Конверсия перехода студентов из одного модуля в другой на каждом курсе', stacked=True)

ax.set_xlabel('Модули курса')
ax.set_ylabel('Конверсия')
ax.set_xticklabels(done_df['title'], rotation=45, ha='right')

fig.tight_layout()
plt.show()
```

```python
# Задание 3

fig, ax = plt.subplots(15, 1, figsize=(15, 70))

courses = np.sort(done_df['title'].unique())

list_colors = ['r', 'g', 'b', 'm']

colors = dict(zip(done_df['field'].unique(), list_colors))

for index, ax in enumerate(ax.flat):
    chart_df = done_df[done_df['title']==courses[index]].sort_values('module_number').reset_index(drop=True)
    chart_df.pivot_table(
        values='conversion', index=chart_df.index, columns='field').plot.barh(
        ax=ax, width=0.7, title=titles[index], color=colors[chart_df['field'].unique()[0]], alpha=0.1)
    ax.vlines(chart_df['conversion'].median(), ax.get_ylim()[0], ax.get_ylim()[1], color='y', linewidth=1)
    ax.set_yticklabels(chart_df.index +1)
    ax.set_xlabel('Конверсия перехода')
    ax.set_ylabel('Модули курса')

fig.tight_layout()
plt.show()
```
